### Examples Loops

```python
data = [53,76,25,98,56,42,69,81]
range(len(data))
```

```python
data_copy = data[:]

for i in range(len(data_copy)):
    for j in range(0, len(data_copy)-i-1):
        #print(j)
        if data_copy[j] > data_copy[j+1]:
            data_copy[j],data_copy[j+1] = data_copy[j+1],data_copy[j]
            
print(data_copy)
            
# Check
# [53, > 76....]
# if yes -> Swtich the values
```
